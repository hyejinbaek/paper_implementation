```python
import matplotlib.pyplot as plt
import numpy as np

def compute_CD(avranks, N, alpha=0.05, test="bonferroni-dunn"):
    k = len(avranks)
    q = np.percentile(range(N), 100 * (1 - alpha / (2 * (k - 1))))
    cd = q / np.sqrt(2) * (k * (k + 1) / (6.0 * N)) ** 0.5
    return cd

def graph_ranks(avranks, names, cd=None, width=6, textspace=1.0):
    y_values = list(range(1, len(names)+1))
    
    # 그림 그리기
    plt.figure(figsize=(10, 6))
    plt.barh(y_values, avranks, color='lightblue', edgecolor='black')
    
    # CD 막대기 표시
    if cd is not None:
        plt.plot([np.mean(avranks)-cd/2, np.mean(avranks)+cd/2], [len(names)+1, len(names)+1], color='red', linewidth=2)
    
    # 그래프 꾸미기
    plt.yticks(y_values, names)
    plt.title('Average Ranks')
    plt.xlabel('Average Rank')
    plt.ylabel('Method')
    
    # 텍스트 표시
    for i, value in enumerate(avranks):
        plt.text(value + textspace, y_values[i], f'{value:.2f}', va='center')
    
    plt.show()

# 정확도 데이터
DataWig = [0.6965, 0.4284, 0.7676, 0.3352, 0.8995, 0.0518, 0.0722, 0.0546, 0.3122, 0.0133, 0.6456, 0.6038, 0.1563, 0.9257, 0.7889, 0.5613, 
           0.8815, 0.4989, 0.4240, 0.0035, 0.0035, 0.3270, 0.5994]
Dynamic = [0.7012, 0.5377, 0.0859, 0.3944, 0.6032, 0.2113, 0.6148, 0.1237, 0.4937, 0.3700, 0.5256, 0.4857, 0.1603, 0.8603, 0.6700, 0.5552,
           0.8833, 0.3883, 0.5925, 0.3677, 0.3677, 0.5016, 0.4825]
Knn = [0.7287, 0.4301, 0.7670, 0.3352, 0.9358, 0.0518, 0.1981, 0.0546, 0.3106, 0.0133, 0.6589, 0.6952, 0.1563, 0.9358, 0.8049, 0.5653, 0.8817,
       0.5411, 0.4240, 0.0035, 0.0035, 0.3270, 0.5994]
Zero = [0.7070, 0.3393, 0.7722, 0.3352, 0.9072, 0.0518, 0.1907, 0.0546, 0.3062, 0.0133, 0.6344, 0.6190, 0.1563, 0.9061, 0.7756, 0.5635,
        0.8751, 0.4783, 0.4240, 0.0035, 0.0035, 0.3270, 0.5994]
proposed = [0.7092, 0.7773, 0.7986, 0.8829, 0.8319, 0.9493, 0.9662, 0.9447, 0.6842, 0.9836, 0.7836, 0.7545, 0.8694, 0.9349, 0.7749,
            0.8293, 0.8686, 0.7023, 0.6348, 0.8846, 0.9956, 0.8466, 0.5604]

avranks = [np.mean(DataWig), np.mean(Dynamic), np.mean(Knn), np.mean(Zero), np.mean(proposed)]

# 계산 및 시각화
N = len(DataWig)
cd = compute_CD(avranks, N)
graph_ranks(avranks, ['DataWig', 'Dynamic', 'Knn', 'Zero', 'proposed'], cd=cd)
```

```python
import pandas as pd
from scipy.stats import friedmanchisquare
from scikit_posthocs import posthoc_conover

# 데이터 로드
df = pd.read_csv('./res/1_breast_ensemble_method_res.csv')
# ± 표기 제거
df['Accuracy'] = df['Accuracy'].str.split(' ±').str[0].astype(float)

# 5개의 메서드 선택
methods_of_interest = ['8_dynamic + datawig + knn', 'datawig', 'dynamic', 'knn', 'zero']
df_selected_methods = df[df['method'].isin(methods_of_interest)]
print(" == df_selected_methods === ", df_selected_methods)

# 프리드먼 테스트
result_friedman = friedmanchisquare(*[df_selected_methods[df_selected_methods['method'] == method]['Accuracy'] for method in methods_of_interest])
print("Friedman Test p-value:", result_friedman.pvalue)

# 본페로니 테스트
posthoc_result = posthoc_conover(df_selected_methods, val_col='Accuracy', group_col='method', p_adjust='bonferroni')
print("\nBonferroni Test Results:")
print(posthoc_result)
```
